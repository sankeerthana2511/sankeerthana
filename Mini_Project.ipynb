```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from transformers import pipeline
from google.colab import files
```

```python
uploaded = files.upload()
video_path = list(uploaded.keys())[0]  # Get the uploaded file's name
```

```python
def extract_keyframes(video_path, num_frames=10):
    cap = cv2.VideoCapture(video_path)
    frame_list = []


    success, frame = cap.read()

    while success:
        frame = cv2.resize(frame, (224, 224))
        frame_list.append(frame)
        success, frame = cap.read()


    frame_array = np.array([frame.flatten() for frame in frame_list])
    kmeans = KMeans(n_clusters=num_frames, random_state=42).fit(frame_array)
    keyframes = [frame_list[i] for i in np.unique(kmeans.labels_)]

    cap.release()
    return keyframes
```

```python
import matplotlib.pyplot as plt

def display_frames(frames):
    plt.figure(figsize=(15, 6))
    for i, frame in enumerate(frames):
        plt.subplot(2, 5, i+1)
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.axis('off')
    plt.show()

display_frames(keyframes)
```
